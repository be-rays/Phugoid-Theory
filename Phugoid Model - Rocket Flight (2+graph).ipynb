```python
import numpy
from matplotlib import pyplot
%matplotlib inline

import math
```

```python
#Start with defining all initial variables

h0=0. #Rocket is at altitude zero at launching
v0=0. #Rocket has velocity zero at launching
ms=50.
g=9.81
rho=1.091
r=0.5
a=math.pi*r**2
ve=325.
cd=0.15
mpo=100.
mp=mpo #At t=0, no propellant has been burnt
```

```python
t=[0.]
step=0.1

dmp=20. #Burn rate between t=0s and t=5s.

h=[]
v=[]
h.append(h0)
v.append(v0)

vt=[] #With vt[m] corresonding to d(v[m])/dt
vt0=-g+dmp*ve/(ms+mp)
vt.append(vt0)

ht=[] #With ht[m] corresonding to d(h[m])/dt
ht0=v0
ht.append(ht0)

print(vt)
```

```python
print (t[0], mpo, v[0])

i=1
while h[i-1]>=0 :
    t.append(t[i-1]+step)
    
    if t[i]>4.95 : 
        mp=0.
        dmp=0
    else :
        dmp=20.
        mp=mpo-dmp*t[i] #As propellant is burnt, we must change the mp variable 
        
    v.append(v[i-1]+step*vt[i-1])
    vt.append(-g+dmp*ve/(mp+ms)-0.5*rho*v[i]*abs(v[i])*a*cd/(mp+ms))
    h.append(h[i-1]+step*v[i-1])
    print (t[i], v[i], h[i])
    
    i=i+1
    

#We observe a minor problem with the time values. What could explain it?

    
```

```python
pyplot.figure(figsize=(10,4))   #set plot size
pyplot.ylim(-400,400)             #y-axis plot limits
pyplot.tick_params(axis='both', labelsize=14) #increase font size for ticks
pyplot.xlabel('t', fontsize=14) #x label
pyplot.ylabel('v', fontsize=14) #y label
pyplot.plot(t,v, 'k-');
```
